```python
catalog_name = "cat_ccondori_24062026"
bronze_name = "bronze"
silver_name = "silver"
gold_name = "gold"
```

```python
spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog_name};")
```

```python
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{bronze_name};")
```

```python
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{silver_name};")
```

```python
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{gold_name};")
```

```python
# 1. Definimos las variables para no repetir código
storage_account = "ccondoristad3"

# 2. Eliminamos los esquemas actuales que se crearon sin ruta
spark.sql(f"DROP SCHEMA IF EXISTS {catalog_name}.{bronze_name}")
spark.sql(f"DROP SCHEMA IF EXISTS {catalog_name}.{silver_name}")
spark.sql(f"DROP SCHEMA IF EXISTS {catalog_name}.{gold_name}")

# 3. Los recreamos asignando la ruta física desde el momento de su creación
spark.sql(f"CREATE SCHEMA {catalog_name}.{bronze_name} MANAGED LOCATION 'abfss://{bronze_name}@{storage_account}.dfs.core.windows.net/'")
spark.sql(f"CREATE SCHEMA {catalog_name}.{silver_name} MANAGED LOCATION 'abfss://{silver_name}@{storage_account}.dfs.core.windows.net/'")
spark.sql(f"CREATE SCHEMA {catalog_name}.{gold_name} MANAGED LOCATION 'abfss://{gold_name}@{storage_account}.dfs.core.windows.net/'")
```
